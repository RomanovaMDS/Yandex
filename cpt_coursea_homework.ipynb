Импортируем библиотеки

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline

# Гамма распределение

```python
gamma = sts.gamma(1)
```

Генерируем выборку объема 1000 элементов

```python
sample = gamma.rvs(1000)
```

```python
x = np.linspace(0,10,100)
```

```python
pdf = gamma.pdf(x)
```

Высчитываем плотность распределения и строим гистограмму

```python
plt.hist(sample, normed=True)
plt.plot(x, pdf)
plt.ylabel('F(x)')
plt.xlabel('x')
plt.show()
```

Строим гистограммы гамма распределения и его выборочных средних

```python
for n in [5, 10, 50]:
    sample_means = []
    for rvn in range(1000):
        sample_means.append(gamma.rvs(n).mean())
    plt.hist(sample_means, normed=True, label='гистограмма')
    norm_rv = sts.norm(1,  np.sqrt(1.*1./n))
    x = np.linspace(0,3,100)
    pdf = norm_rv.pdf(x)
    plt.plot(x, pdf, label='распределение плотности')
    plt.legend()
    plt.title("Размер выборки %d" % n)
    plt.ylabel('f(x)')
    plt.xlabel('x')
    plt.show()
```

Чем больше размер выборки, тем лучшее сходится распределение средних значений.
